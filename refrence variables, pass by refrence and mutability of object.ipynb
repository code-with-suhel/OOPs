# How objects access attributes

```python
class Person:
    def __init__(self,name_input,country_input):
        self.name = name_input
        self.country = country_input
    
    def greet(self):
        if self.country == 'india':
            print('Namaste',self.name)
        else:
            print('Hello',self.name)
```

```python
p = Person("nitish", "india")
```

```python
# how to access attributes
p.name
```

```python
p.country
```

```python
# how to access methods
p.greet()
```

```python
# what if i try to access non-existent attributes
p.gender
```

# Reference Variables

- Reference variables hold the objects
- We can create objects without reference variable as well
- An object can have multiple reference variables
- Assigning a new reference variable to an existing object does not create a new object

### Remark:

##### remeber that, technically p is not the object of the class, it is refrence varibale which contains the address/refrence of the object 

```python
# object without a reference
class Person:
    def __init__(self):
        self.name = 'nitish'
        self.gender = 'male'
Person()
```

```python
p = Person()
q = p # both q and p points to Person object. --> we can give the object addres to multipe refrence variables. 
```

```python
# Multiple ref
print(id(p))
print(id(q))
```

```python
# change attribute value with the help of 2nd object
print(p.name)
print(q.name)
q.name = 'ankit'
print(q.name)       # when q changes p was also changed.
print(p.name)
```

# Pass by reference

```python
class Person:
    
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender

# outside the class -> function
def greet(person):
    print("hi my name is",person.name,"i am a",person.gender)
    
p = Person('nitish','male')
greet(p)   # we pass the object in func.
```

```python
# fuction returns the object.
class Person:
    
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender

# outside the class -> function
def greet(person):
    print("hi my name is",person.name,"i am a",person.gender)
    p1 = Person("ankit", "male")
    return p1
    
p = Person('nitish','male')
x = greet(p) 
print(x.name)
print(x.gender)
```

```python
# hence when we pass the object to the function, actually we pass the refrence variable of the object not the object..
# and that is why we call it "pass by refrence".
```

```python
# therefore, pass by refrence = when we pass the object(its refrence) as an argument to the fuction.
```

```python
class Person:

  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

# outside the class -> function
def greet(person):
  print(id(person))
  person.name = 'ankit'
  print(person.name)

p = Person('nitish','male')
print(id(p))
greet(p)
print(p.name)
```

# Object ki mutability

```python
class Person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender

# outside the class -> function
def greet(person):
    person.name = 'ankit'
    return person

p = Person('nitish','male')
print(id(p))
p1 = greet(p)
print(id(p1))
```
